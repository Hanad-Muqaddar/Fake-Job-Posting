```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('job_train.csv')
```

```python
df.dropna(subset=['description'], axis=0, inplace=True)
df['location'] = df.location.fillna('not specified')
df['requirements'] = df.requirements.fillna('not specified')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer()
```

```python
df["requirements"] = vect.fit_transform(df.requirements).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
```

```python
df["title"] = vect.fit_transform(df.title).toarray()
```

```python
df[700:900]
```

```python
def vectorize_data(column):
    df[column] = vectorizer.fit_transform(df[column])
text_columns = ['location','description']
```

```python
for column in text_columns:
    vectorize_data(column)
```
